```python
import pandas as pd
import numpy as np
from keras.utils import to_categorical
from sklearn.model_selection import train_test_split
from keras.models import Model
from keras.layers import Dense, Input
from sklearn import metrics
```

### 2. Load the data

```python
data_file = pd.read_csv('data.csv', index_col=0)
labels_file = pd.read_csv('labels.csv', index_col=0)
```

```python
data_file
```

```python
labels_file
```

### 4. Merge the labels and the gene expression data

```python
df = data_file.join(labels_file['Class'])
df
```

### 5. Split data (features and labels)

```python
# Features
features = df.iloc[:, :-1] # lines, colums

# Labels
labels = df.iloc[:, -1]
```

```python
print(features)
print('shape -> ', features.shape)
```

```python
print(labels)
print('labels -> ', labels.shape)
```

### 6. Re-encode class column
**Piece of code to re-encode the labels with integers**

```python
cancer_col = df['Class'].to_numpy()  # Get the cancer type column
cancer_encoding_dict = {}  # Initialise the dictionary of cancer type associated to an int
cancer_int_list = []  # Initialise the list with the serie of int replacing cancer type
new_int_value = 0  # Int value

for cancer in cancer_col:

    # If the cancer is not in the dictionary insert it
    if cancer not in cancer_encoding_dict:
        cancer_encoding_dict[cancer] = new_int_value
        new_int_value += 1  # Iter the integer

    # Append the list with the integer
    cancer_int_list.append(cancer_encoding_dict[cancer])

print(cancer_encoding_dict)
```

### 7. Neural Network

#### 7.1 Split data

```python
# Create an array with categories in binary
labels_bis = to_categorical(cancer_int_list)
# Split the data in train and test for features and labels
features_train, features_test, labels_train, labels_test = train_test_split(features, labels_bis, test_size=0.33, random_state=42, stratify=labels_bis)
```

#### 7.2 Create layers

```python
# Define layers of the Neural Network
init = 'random_uniform'  # Kernel initializer : statistical distribution to initialise weights of the NN

input_layer = Input(shape=(20531,))
mid_layer = Dense(15, activation = 'relu', kernel_initializer = init)(input_layer)
mid_layer_2 = Dense(8, activation = 'relu', kernel_initializer = init)(mid_layer)
output_layer = Dense(5, activation = 'softmax', kernel_initializer = init)(mid_layer_2)
```

#### 7.3 Group, Compile, Summarize

```python
# Groups layers into an object Model
model = Model(inputs = input_layer, outputs = output_layer)
# Compile the model and define parameters like optimizer, loss function, metrics for prediction
model.compile(optimizer='sgd',loss='binary_crossentropy',metrics=['accuracy'])
# Display a summary
model.summary()
```

#### 7.4 Fitting the model

The [Fit function](https://keras.io/api/models/model_training_apis/) is used here.

**x** is the input data.

**y** the target data.

**batch** size set as default which is the number of samples per gradient update.

**epoch** is the number of epochs to train the model i.e. an iteration over the entire x and y data provided.

**verbode** set on the default mode. Verbosity mode. 0 = silent, 1 = progress bar, 2 = one line per epoch.

```python
model.fit(features_train, labels_train, batch_size=32, epochs=50, verbose=1)
```

### 8. Model prediction

```python
prediction = model.predict(features_test)
```

```python
prediction
```

### 9. Evaluate the model
[Argmax](https://www.kite.com/python/docs/keras.backend.argmax) is an operation that finds the argument that gives the maximum value from a target function.

```python
np.argmax(prediction, axis=1), np.argmax(labels_test, axis=1)
```

[Accuracy](https://keras.io/api/metrics/accuracy_metrics/) of the test

```python
metrics.accuracy_score(np.argmax(labels_test, axis = 1),np.argmax(prediction, axis = 1))
```

Create a crosstab (a table) from labels of the test and the prediction in argmax

```python
pd_res = pd.crosstab(np.argmax(labels_test, axis = 1),np.argmax(prediction, axis = 1))
```
